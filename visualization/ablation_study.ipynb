```python
import matplotlib.pyplot as plt
import os

# ---------- directories ----------
fig_dir = './figures'
os.makedirs(fig_dir, exist_ok=True)

# ---------- shared x-axis ----------
k_vals = [2, 3, 5, 6, 8, 10, 20, 30, 40]

# ---------- Spider ----------
recall_spider = {
    "GRASTSQL 0.6B": [0.81, 0.918, 0.984, 0.99, 0.996, 0.998, 1, 1, 1],
    "w/o Steiner Tree":                 [0.7272, 0.8627, 0.962, 0.977, 0.9903, 0.9953, 0.9996, 0.9998, 1],
    "w/o Func. Dep. Graph":           [0.7137, 0.8438, 0.9469, 0.9659, 0.9865, 0.9939, 0.9998, 1, 1],
}

precision_spider = {
    "GRASTSQL 0.6B": [0.701, 0.648, 0.471, 0.404, 0.316, 0.263, 0.17, 0.155, 0.151],
    "w/o Steiner Tree":                 [0.8337, 0.71, 0.5036, 0.4315, 0.3342, 0.2741, 0.1726, 0.1556, 0.1509],
    "w/o Func. Dep. Graph":           [0.814, 0.6862, 0.4911, 0.4231, 0.3317, 0.2734, 0.1726, 0.1556, 0.1509],
}


# ---------- BIRD ----------
recall_bird = {
    "GRASTSQL 0.6B": [0.579, 0.745, 0.889, 0.919, 0.949, 0.965, 0.992, 0.997, 0.999],
    "w/o Steiner Tree":      [0.4741, 0.6271, 0.8282, 0.8779, 0.9286, 0.953, 0.9906, 0.9963, 0.9987],
    "w/o Func. Dep. Graph": [0.4705, 0.6107, 0.7957, 0.8463, 0.9033, 0.9329, 0.984, 0.9939, 0.9966],
}

# Precision values for BIRD (Encoder variants)
precision_bird = {
    "GRASTSQL 0.6B": [0.811, 0.761, 0.648, 0.58, 0.464, 0.383, 0.219, 0.162, 0.135],
    "w/o Steiner Tree":      [0.8669, 0.7949, 0.6642, 0.5954, 0.4806, 0.398, 0.224, 0.1636, 0.1359],
    "w/o Func. Dep. Graph": [0.8567, 0.7723, 0.6335, 0.5696, 0.4646, 0.3875, 0.222, 0.1632, 0.1356],
}

# for Spider , k_vals[:-3], fort BIR

# 
# k_vals = k_vals[:-2]
for metric in [recall_spider, precision_spider, recall_bird, precision_bird]:
    for k in metric:
        metric[k] = metric[k][:-2]

# ---------- helper ----------
def plot_metric(ax, data, ylabel, title, ylim):
    _k_vals = k_vals
    for label, vals in data.items():
        print(label, vals)
        if "Spider" in title:
            vals = vals[:6]
            _k_vals = _k_vals[:6]
        else:
            vals = vals[:7]
            _k_vals = _k_vals[:7]
        
        ax.plot(_k_vals, vals, marker='o', label=label)
    ax.set_xlabel('k', fontsize=14)
    ax.set_ylabel(ylabel, fontsize=14)
    ax.set_title(title, fontsize=14)
    ax.set_xticks(_k_vals)
    ax.tick_params(axis='x', labelsize=14)  # Increase xticks font size
    ax.set_ylim(*ylim)
    ax.grid(True)

# ---------- 4-panel figure ----------
fig, axes = plt.subplots(1, 4, figsize=(16, 3.5))

plot_metric(axes[0], recall_spider, 'Recall',    'Recall  Spider',   (0.6, 1.05))
plot_metric(axes[1], precision_spider, 'Precision','Precision  Spider',(0.05, 0.85))
plot_metric(axes[2], recall_bird,    'Recall',    'Recall  BIRD',    (0.4, 1.05))
plot_metric(axes[3], precision_bird, 'Precision','Precision  BIRD', (0.05, 0.9))

# reserve space for the shared legend
plt.tight_layout(rect=(0, 0.15, 1, 1))

# adjust each axis position (shrink height)
for ax in axes:
    box = ax.get_position()
    ax.set_position([box.x0, box.y0, box.width, box.height*0.8])

# single horizontal legend beneath all plots
handles, labels = axes[0].get_legend_handles_labels()
# fig.legend(handles, labels,
#            loc='lower center', ncol=3, frameon=False, bbox_to_anchor=(0.5, 0.1))
fig.legend(handles, labels,
           loc='lower center',
           ncol=3,
           frameon=False,
           fontsize=14,
           bbox_to_anchor=(0.5, 0.06))          #  font size in points

# save
plt.savefig(os.path.join(fig_dir, 'ablation_four_panel.png'), dpi=300)
plt.show()
```
